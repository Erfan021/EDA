## Transpose a `Panel` with the `.transpose()` Method

```python
companies = ["MSFT", "GOOG", "AAPL", "IBM", "AMZN"]

p = data.DataReader(name = companies, data_source = "yahoo", start = "2010-01-01", end = "2016-12-31")

p
```

```python
p.axes
```

```python
p2 = p.transpose(2, 1, 0)
```

```python
p
```

```python
p2["GOOG"]
```

```python
p2.major_xs("2010-01-04")
```

```python
p2.minor_xs("Volume")
```

## The `.swapaxes()` Method

```python
companies = ["MSFT", "GOOG", "AAPL", "IBM", "AMZN"]

p = data.DataReader(name = companies, data_source = "yahoo", start = "2010-01-01", end = "2016-12-31")

p
```

```python
p2 = p.swapaxes("items", "minor")
```

```python
p2.axes
```

```python
p2["MSFT"]
p2.major_xs("2016-09-02")
p2.minor_xs("Close")
```
